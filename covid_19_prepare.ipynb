1. **Country/Region**: Quốc gia hoặc khu vực mà dữ liệu COVID-19 đang được báo cáo.

2. **Confirmed**: Tổng số ca nhiễm COVID-19 đã được xác nhận tại quốc gia hoặc khu vực đó.

3. **Deaths**: Tổng số ca tử vong do COVID-19 tại quốc gia hoặc khu vực đó.

4. **Recovered**: Tổng số ca đã hồi phục sau khi nhiễm COVID-19.

5. **Active**: Số ca hiện tại vẫn đang bị nhiễm COVID-19 (tính bằng: Confirmed - Deaths - Recovered).

6. **New cases**: Số ca nhiễm COVID-19 mới được báo cáo trong ngày.

7. **New deaths**: Số ca tử vong mới do COVID-19 trong ngày.

8. **New recovered**: Số ca hồi phục mới trong ngày.

9. **Deaths / 100 Cases**: Tỷ lệ phần trăm số ca tử vong trên tổng số ca nhiễm (tính bằng: Deaths / Confirmed * 100).

10. **Recovered / 100 Cases**: Tỷ lệ phần trăm số ca hồi phục trên tổng số ca nhiễm (tính bằng: Recovered / Confirmed * 100).

11. **Deaths / 100 Recovered**: Tỷ lệ phần trăm số ca tử vong trên số ca hồi phục (tính bằng: Deaths / Recovered * 100).

12. **Confirmed last week**: Tổng số ca nhiễm đã được xác nhận vào tuần trước.

13. **1 week change**: Sự thay đổi về số ca nhiễm mới trong tuần qua (tính bằng: Confirmed - Confirmed last week).

14. **1 week % increase**: Phần trăm tăng số ca nhiễm trong tuần qua (tính bằng: (1 week change / Confirmed last week) * 100).

15. **WHO Region**: Khu vực của Tổ chức Y tế Thế giới (WHO) mà quốc gia hoặc khu vực đó thuộc về. Ví dụ: Khu vực Tây Thái Bình Dương, Châu Âu, Châu Phi, Châu Mỹ, Đông Địa Trung Hải, hoặc Đông Nam Á.

1. **Province/State**: Tên tỉnh hoặc bang của quốc gia mà dữ liệu COVID-19 được ghi nhận. Với một số quốc gia nhỏ hoặc không phân chia hành chính, cột này có thể để trống (NaN).

4. **Long**: Kinh độ (Longitude) của quốc gia hoặc khu vực, cũng dùng để định vị trên bản đồ địa lý.

5. **Date**: Ngày ghi nhận dữ liệu cụ thể về số ca nhiễm, tử vong, hồi phục, và các thông tin khác liên quan đến COVID-19.

6. **Confirmed**: Tổng số ca nhiễm COVID-19 đã được xác nhận tính đến ngày cụ thể.

7. **Deaths**: Tổng số ca tử vong do COVID-19 được ghi nhận đến ngày đó.

8. **Recovered**: Tổng số ca đã hồi phục sau khi nhiễm COVID-19 tính đến ngày cụ thể.

9. **Active**: Số ca hiện tại vẫn đang nhiễm COVID-19, thường được tính bằng công thức: `Active = Confirmed - Deaths - Recovered`.

10. **WHO Region**: Khu vực của Tổ chức Y tế Thế giới (WHO) mà quốc gia hoặc khu vực đó thuộc về. WHO phân chia các quốc gia vào các khu vực như Châu Mỹ, Châu Âu, Tây Thái Bình Dương, Châu Phi, Đông Nam Á, và Đông Địa Trung Hải.

# Worldometer

Dưới đây là ý nghĩa của các trường thông tin trong dữ liệu COVID-19:

1. **Country/Region**: Tên quốc gia hoặc khu vực mà dữ liệu COVID-19 được ghi nhận.

2. **Continent**: Châu lục mà quốc gia hoặc khu vực đó thuộc về.

3. **Population**: Tổng số dân số của quốc gia hoặc khu vực.

4. **TotalCases**: Tổng số ca nhiễm COVID-19 đã được xác nhận tại quốc gia hoặc khu vực đó.

5. **NewCases**: Số ca nhiễm COVID-19 mới được báo cáo trong một khoảng thời gian cụ thể (thường là trong ngày).

6. **TotalDeaths**: Tổng số ca tử vong do COVID-19 tại quốc gia hoặc khu vực đó.

7. **NewDeaths**: Số ca tử vong mới do COVID-19 trong một khoảng thời gian cụ thể (thường là trong ngày).

8. **TotalRecovered**: Tổng số ca đã hồi phục sau khi nhiễm COVID-19 tại quốc gia hoặc khu vực đó.

9. **NewRecovered**: Số ca hồi phục mới trong một khoảng thời gian cụ thể (thường là trong ngày).

10. **ActiveCases**: Số ca hiện tại vẫn đang nhiễm COVID-19 (tính bằng: `TotalCases - TotalDeaths - TotalRecovered`).

11. **Serious,Critical**: Số ca bệnh COVID-19 đang ở trạng thái nghiêm trọng hoặc nguy kịch.

12. **Tot Cases/1M pop**: Số ca nhiễm COVID-19 tổng cộng trên mỗi triệu dân (tính bằng: `TotalCases / Population * 1,000,000`).

13. **Deaths/1M pop**: Số ca tử vong do COVID-19 trên mỗi triệu dân (tính bằng: `TotalDeaths / Population * 1,000,000`).

14. **TotalTests**: Tổng số xét nghiệm COVID-19 đã được thực hiện tại quốc gia hoặc khu vực đó.

15. **Tests/1M pop**: Số xét nghiệm COVID-19 trên mỗi triệu dân (tính bằng: `TotalTests / Population * 1,000,000`).

16. **WHO Region**: Khu vực của Tổ chức Y tế Thế giới (WHO) mà quốc gia hoặc khu vực đó thuộc về, ví dụ: Châu Âu, Châu Mỹ, Tây Thái Bình Dương, Châu Phi, Đông Nam Á, hoặc Đông Địa Trung Hải.

Các trường thông tin này cung cấp cái nhìn tổng quan và chi tiết về tình hình dịch bệnh COVID-19 ở từng quốc gia hoặc khu vực, cũng như cách dịch bệnh ảnh hưởng đến dân số và hệ thống y tế.
